```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

SELECT A LARGE REAL WORLD DATASET FROM KAGGLE

```python
# Opening csv file using pandas
df = pd.read_csv("tracks.csv")
df.head()
```

PERFORM DATA PREPARATION AND CLEANING USING PANDAS & NUMPY

```python
# Checking for null values
pd.isnull(df).sum()
```

```python
# Information regarding dataframe
df.info()
```

```python
# Dropping irrelevent columns from dataset
df = df.drop(["key", "mode", "explicit"], axis=1)
df.head()
```

```python
# Converting duration of songs from millisecond into seconds
df["duration"] = df["duration_ms"].apply(lambda x: round(x/1000))
df.drop("duration_ms", inplace=True, axis = 1)
df.head()
```

ASK & ANSWER QUESTIONS ABOUT THE DATA

```python
# Five most popular song present in Spotify dataset
most_popular_songs = df.sort_values('popularity', ascending=False).head()
most_popular_songs
```

```python
# Five least popular song present in Spotify dataset
least_popular_song = df.sort_values('popularity', ascending=True).head()
least_popular_song
```

```python
# Song with the longest duration
longest_duration = df.sort_values('duration', ascending=False).head(1)
longest_duration
```

```python
# Song with the shortest duration
shortest_duration = df.sort_values('duration', ascending=True).head(1)
shortest_duration
```

PERFORM EXPLORATORY ANALYSIS AND VISUALIZATION USING MATPLOTLIB & SEABORN

```python
# Correlation Heatmap between variables
df = df.corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(df, annot=True, fmt=".1g", vmin=-1, vmax = 1, center = 0, cmap = "Spectral", linewidth=1, linecolor = "Black")
heatmap.set_title("Correlation Heatmap between variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
# Directly proportional relation between Loudness and Energy
%matplotlib inline
plt.figure(figsize=(10,6))
plt.title("Directly proportional relation between Loudness and Energy")
plt.xlabel("energy")
plt.ylabel("loudness")
plt.scatter(df.energy, df.loudness, color = 'c')
```

```python
# Correlation between Loudness and Energy
plt.figure(figsize=(10,6))
sns.regplot(data = df, y = 'loudness', x= 'energy', color='c').set(title = 'Loudness vs Energy Correlation')
```

```python
# Directly proportional relation between Acousticness and Instrumentalness
%matplotlib inline
from turtle import color


plt.figure(figsize=(10,6))
plt.title("Directly proportional relation between Acousticness and Instrumentalness")
plt.xlabel("instrumentalness")
plt.ylabel("acousticness")
plt.scatter(df.instrumentalness, df.acousticness, color='b')
```

```python
# Acousticness vs Instrumentalness Correlation
plt.figure(figsize=(10,6))
sns.regplot(data = df, y = 'acousticness', x= 'instrumentalness', color='b').set(title = 'Acousticness vs Instrumentalness Correlation')
```
